```python
import pandas as pd

# Load the Excel file into a DataFrame
data = pd.read_excel("Party_Criminal_Cases.xlsx")

# Group the data by party name and count the number of criminal cases for each party
party_criminal_cases = data.groupby('Party')['Criminal Cases'].max()

print("Maximum criminal cases for each party:")
print(party_criminal_cases)

```

```python
# find mean of criminal cases
mean_Criminal_Cases = data['Criminal Cases'].mean()
print("Mean Criminal Cases:", mean_Criminal_Cases)
```

```python
#find variance of criminal cases
variance_Criminal_Cases = data['Criminal Cases'].var()
print("Variance of Criminal Cases:", variance_Criminal_Cases)
```

```python
# find standard deviation(SD) of criminal cases
standard_deviation_Criminal_Cases = data['Criminal Cases'].std()
print("Standard Deviation of Criminal Cases:", standard_deviation_Criminal_Cases)
```

```python
for index, row in data.iterrows():
    print('Row:', index, 'Criminal Cases:', row['Criminal Cases'])
```

```python
# Calculate the Percentage of 'Criminal cases' column
total_Criminal_Case = data['Criminal Cases'].sum()
Percentage_Criminal_Cases = (data['Criminal Cases'] / total_Criminal_Case) * 100
print(Percentage_Criminal_Cases)
```

```python
# find the candidate name which have more total assets
# Assuming 'Candidate' is the column containing candidate names and 'Total Assets' is the column containing their total assets
candidate_with_max_assets = data.groupby('Candidate')['Total Assets'].sum().sort_values(ascending=True)


print("Candidate with the highest total assets:", candidate_with_max_assets)
```

```python
# Find the maximum value in the 'Criminal cases' column
max_criminal_cases = data['Criminal Cases'].max()
print("Max Criminal Cases:", max_criminal_cases)
```

```python
# Find the minimum value in the 'Criminal cases' column
min_criminal_cases = data['Criminal Cases'].min()
print("Min Criminal Cases:", min_criminal_cases)
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
data = pd.read_excel("Party_Criminal_Cases.xlsx")

# Convert 'Criminal Cases' column to numeric type if it's not already
data['Criminal Cases'] = pd.to_numeric(data['Criminal Cases'], errors='coerce')

# Group the data by party name and calculate the total criminal cases for each party
party_criminal_cases = data.groupby('Party')['Criminal Cases'].sum()

# Plotting
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
party_criminal_cases.plot(kind='bar', color='skyblue')
plt.title('Total Criminal Cases by Party')
plt.xlabel('Party')
plt.ylabel('Total Criminal Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```
